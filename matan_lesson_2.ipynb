### **6.** Прочитайте высказывания, установите их истинность и постройте противоположное высказывание:

$a)\,\,\, \forall x\in\mathbb{R}\,\,\,\exists X\in\mathbb{R}:\,\,\, X>x;$

$б)\,\,\, \forall y\in\Bigl[0; \frac{\pi}{2}\Bigr]\,\,\,\exists \varepsilon>0:\,\,\, \sin y<\sin(y+\varepsilon);$

$в)\,\,\, \forall y\in\Bigl[0; \pi\Bigr)\,\,\,\exists \varepsilon>0:\,\,\, \cos y>\cos(y+\varepsilon).$




#### a) Для любого рационального числа $x$ найдется большее по значению число $X$, также принадлежащее множеству рациональных чисел.

Противоположное высказывание: $ \exists x\in\mathbb{R}\,\,\,\forall X\in\mathbb{R}:\,\,\, X \leqslant x;$ У рационального множества есть максимум... (существует рац. число, такое что, все числа множества $\mathbb{R}$ меньше его или равны ему).

Противоположное утверждение ложно, значит исходное истинно.


#### б)  Для любого $\boldsymbol y$, принадлежащего замкнутому интервалу от 0 до $ \boldsymbol{ \pi / 2}$, найдется такое положительное число $\boldsymbol \epsilon$, что $ \sin y<\sin(y+\varepsilon)$.

Противоположное высказывание: $\exists y\in\Bigl[0; \frac{\pi}{2}\Bigr]\,\,\,\forall \varepsilon>0:\,\,\, \sin y \geqslant \sin(y+\varepsilon);$ На указанном интервале $\sin$ возрастает, значит противоположное высказывание верно (достаточно было бы одной точки в которой оно выполняется, но здесь - верно во всех, кроме одной точки).

Противоположное высказывание истинно, значит исходное - ложно.


#### в)  Для любого $\boldsymbol y$, принадлежащего полу-замкнутому интервалу от 0 до $ \boldsymbol{ \pi}$($ \boldsymbol{ \pi}$ - не входит), найдется такое положительное число $\boldsymbol \epsilon$, что $ \cos y<\cos(y+\varepsilon)$.

Это высказывание истинно.

Противоположное высказывание: 

$ \exists y\in\Bigl[0; \pi\Bigr)\,\,\,\forall \varepsilon>0:\,\,\, \cos y\leqslant\cos(y+\varepsilon).$
Это выражение ложно, т.к. найдутся такие $\varepsilon>0$, при которых неравенство не выполняется.
